```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(10)
y = np.e ** x
```

```python
plt.figure(figsize=(10, 6))

plt.plot(x, y, label='e**x')
plt.legend(loc='upper center')

plt.grid(axis='y')

plt.show()
```

```python
data = np.random.normal(0, 1, (1000000))
```

```python
plt.hist(data, bins=100)
plt.show()
```

```python
season = ['sp', 'su', 'fa', 'wi']
values = [100, 110, 120, 90]
```

```python
plt.figure(figsize=(8, 6))
plt.bar(season, values)

plt.title('1403 sale')

plt.xlabel('season')
plt.ylabel('value')

plt.show()
```

```python
plt.boxplot(data)
plt.show()
```

```python
plt.pie(values, labels=season)
plt.show()
```

```python
xs = np.random.normal(0, 1, (100))
ys = np.random.normal(0, 1, (100))

plt.scatter(xs, ys)
plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.plot(x, y, label='e**x')
plt.plot(x, y*2, label='e**x2')

plt.legend(loc='upper center')

plt.grid(axis='y')

plt.show()
```

```python
fig, ax1 = plt.subplots(2, 3, figsize=(12, 8))

ax1[0, 0].plot(x, y)
ax1[0, 1].bar(season, values)
ax1[0, 2].scatter(xs, ys)


fig.tight_layout()
```

```python
datasets = pd.read_csv("https://raw.githubusercontent.com/vincentarelbundock/Rdatasets/master/datasets.csv")
main_df = pd.read_csv(datasets.iloc[1046]["CSV"])
main_df['date'] = pd.to_datetime(main_df['date'])
main_df = main_df.assign(
    year_month = lambda x: x.date.dt.to_period('M')
)
df = main_df.pivot_table(
    values='id',
    columns='year_month',
    index='name',
    aggfunc='nunique'
)
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(15, 10))
sns.heatmap(df)
plt.xlabel('YearMonth')
plt.ylabel('ProductName')
plt.show()
```
